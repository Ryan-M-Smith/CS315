```python
A = imread("CS315/HW08/landscape.jpg")
X = np.mean(A,-1);
```

```python
img = plt.imshow(X)
img.set_cmap("gray")
plt.axis("off")
plt.show()
```

```python
U, S, VT = np.linalg.svd(X,full_matrices=False)
S=np.diag(S)
```

```python
k = 20
Xapprox = U[:,:k] @ S[0:k,:k] @ VT[:k,:]
plt.figure(1)
img=plt.imshow(Xapprox)
img.set_cmap("gray")
plt.axis("off")
plt.title(f"k = {str(k)}")
plt.show()
```
