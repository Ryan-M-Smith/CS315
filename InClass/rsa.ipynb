# RSA Demo

```python
import math, random
```

## Calculate $e$, $d$, and $\phi$

```python
p, q = 23, 29
n = p * q
phi = (p - 1) * (q - 1)
e = random.randrange(1, phi)

while math.gcd(e, phi) != 1:
    e = random.randrange(1, phi)

print(f"ðœ™ = {phi}")
print(f"e = {e}")
```

```python
def mod_inverse(e: int, phi: int) -> int:
  e %= phi

  for x in range(1, phi):
    if (e * x) % phi == 1:
      return x

  return 1
```

```python
d = mod_inverse(e, phi)
print(f"d = {d}")
```

## Encrypt and Decrypt a Message

```python
# Encryption

message = [104, 101, 121] # Plaintext: hey
c: list[int] = []
for num in message:
  c.append((int(num) **e ) % n)

print(f"Ciphertext: {c}")
```

```python
# Decryption

plaintext: list[chr] = []
for num in c:
  plaintext.append(chr((int(num) ** d) % n))

word = " ".join(plaintext)
print(word)
```
