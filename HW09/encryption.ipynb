```python
#
# Solve a Caesar cipher
#

ciphertext = [104, 105, 97, 108, 108]
plaintext = list(map(chr, ciphertext))
print(" ".join(plaintext))
```

```python
#
# Solve a Caesar cipher with a shift
#

shift = 3
ciphertext = [107, 104, 124, 100, 111, 111]
shifted = list(map(lambda x: x - shift, ciphertext))
plaintext = list(map(chr, shifted))
print(" ".join(plaintext))
```

```python
input_str = "heyall"

# Calculate the next perfect square size for the matrix
current_len = len(input_str)
size = int(math.ceil(math.sqrt(current_len)))

# Pad the string to the calculated perfect square length
target_len = size * size
padding_needed = target_len - current_len
padded = input_str + 'x' * padding_needed # Using 'x' as padding character

letters = [l for l in padded]
letters = list(map(ord, letters))

# Calculate
A = np.random.randint(low=0, high=5, size=(size, size))
B = np.array(letters).reshape(size, size)
C = np.dot(A, B)

print(f"A = {A}")
print(f"B = {B}")
print(f"C = {C}")
```
